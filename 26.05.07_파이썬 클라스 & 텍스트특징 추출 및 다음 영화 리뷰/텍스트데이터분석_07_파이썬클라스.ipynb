파이썬 클래스

```python
class Account:
    def __init__(self, balance=0):
        # 인스턴스가 생성될 때 각자의 잔액을 가집니다.
        self.balance = balance

    def deposit(self, amount):
        # 입금
        self.balance += amount
        
    def withdraw(self, amount):
        # 잔액이 부족할 때는 출금이 안 되도록 조건문 추가
            self.balance -= amount
        
# 객체 생성 및 실행
a1 = Account(100)  # 초기 잔액 100원으로 시작
a1.deposit(100)    # 100원 입금 -> 200원
a1.withdraw(50)    # 50원 출금 -> 150원
a1.withdraw(300)   # 잔액 부족 메시지 출력

print('최종 잔액 a1 :', a1.balance)
```
