### 2. 서울시 거주/비거주 고객 비교 분석
#### 2-1.고객수, 총 소비액, 카드 이용 건수

```python

bccard_df = pd.DataFrame(bccard)
A=len(bccard_df[bccard_df.CSTMR_MEGA_CTY_NM == '서울특별시']) #고객 수를 len 으로 서울 시 거주 조건은 boolean 안덱스 조건 식으로 요청
B=len(bccard_df[bccard_df.CSTMR_MEGA_CTY_NM != '서울특별시']) #고객 수를 len 으로 서울 시 비 거주 조건은 boolean 안덱스 조건 식으로 요청
print(f'서울시 거주:{A},서울시 外 거주:{B}')
```

#### 총 소비액

```python
bccard_df = pd.DataFrame(bccard)
A=bccard_df[bccard_df.CSTMR_MEGA_CTY_NM == '서울특별시']['AMT'].sum() #총 합을 Sum 으로, 서울 시 거주 조건은 boolean 안덱스 조건 식으로, 소비액은 AMT로 요청
B=bccard_df[bccard_df.CSTMR_MEGA_CTY_NM != '서울특별시']['AMT'].sum() #총 합을 Sum 으로, 서울 시  미 거주 조건은 boolean 안덱스 조건 식으로, 소비액은 AMT로 요청
print(A,B)
```

#### 카드 이용 건 수

```python
bccard_df = pd.DataFrame(bccard)
A=bccard_df[bccard_df.CSTMR_MEGA_CTY_NM == '서울특별시']['CNT'].sum() # 위 사항과 거의 동등 하되 카드 이용 건 수는 CNT로 표현
B=bccard_df[bccard_df.CSTMR_MEGA_CTY_NM != '서울특별시']['CNT'].sum() # 위 사항과 거의 동등 하되 카드 이용 건 수는 CNT로 표현
print(A,B)
```
